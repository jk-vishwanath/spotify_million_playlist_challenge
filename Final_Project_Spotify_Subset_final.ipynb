```python
import pandas as pd
import string
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download stopwords if not already available
#nltk.download('stopwords')
#nltk.download('punkt')

# Load English stop words
stop_words = set(stopwords.words('english'))

# Define a function to preprocess the lyrics and create a bag of words
def create_bag_of_words(lyrics):
    if pd.isna(lyrics):  # Handle missing lyrics
        return None

    # Convert to lowercase
    lyrics = lyrics.lower()

    # Remove punctuation
    lyrics = lyrics.translate(str.maketrans('', '', string.punctuation))

    # Tokenize the text into words
    words = word_tokenize(lyrics)

    # Remove stop words
    filtered_words = [word for word in words if word not in stop_words and word.isalpha()]

    # Count the frequency of each word
    word_counts = Counter(filtered_words)

    # Get the 10 most common words
    most_common_words = [word for word, count in word_counts.most_common(10)]

    # Return as a space-separated string (bag of words)
    return ' '.join(most_common_words)


df['bag_of_words'] = df['lyrics'].apply(create_bag_of_words)

# Display the resulting DataFrame with the new 'bag_of_words' column
df[['track_track_name', 'lyrics', 'bag_of_words']].head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.sentiment import SentimentIntensityAnalyzer

# Sample DataFrame setup (Assuming df already exists with 'lyrics' and 'bag_of_words' columns)
# Example: df = pd.DataFrame({'lyrics': [...], 'bag_of_words': [...]})

# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Apply sentiment analysis on the 'bag_of_words' column
df['sentiment_bag_of_words'] = df['bag_of_words'].fillna('').apply(lambda words: sia.polarity_scores(words)['compound'])

# Initialize TF-IDF Vectorizer for bag_of_words
tfidf_vectorizer = TfidfVectorizer(max_features=500, stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['bag_of_words'].fillna(''))

# Convert to DataFrame for analysis
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Combine with original DataFrame
df = pd.concat([df.reset_index(drop=True), tfidf_df.reset_index(drop=True)], axis=1)

# Display the updated DataFrame with sentiment and TF-IDF features
df[['track_track_name', 'bag_of_words', 'sentiment_bag_of_words']].head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualization of sentiment scores
plt.figure(figsize=(10, 6))
sns.histplot(df['sentiment_bag_of_words'], bins=20, kde=True)
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating a sentiment category based on sentiment score
def categorize_sentiment(score):
    if score > 0.05:
        return 'positive'
    elif score < -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment_category'] = df['sentiment_bag_of_words'].apply(categorize_sentiment)
```
